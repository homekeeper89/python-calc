```python
from tkinter import *

window = Tk()
window.title("my test");
# 1. 레이블 : 윈도우에 텍스트를 표시할때 주로 사용.
Label(window, text = "정의되어 있는 단어를 입력하고 엔터키를 누르세요").grid(row = 0, column = 0, sticky = W)
# w 는 서쪽을 의미함

# 2. 엔트이 위젯 : 텍스트를 입력할 수 있는 상자를 제공
# 먼저 엔트리 객체를 생성 후에 위치를 조절하는 것이 좋다
entry = Entry(window, width = 20, bg = "light blue");
entry.grid(row = 1, column = 0, sticky = W)

# 3. 버튼 위젯 :버튼을 눌러을때 동작할 함수를 먼저 작성 한 후에 버튼을 만들고 함수를 넣어준다.
def click():
    text = entry.get(); # entry로 부터 얻은 텍스트
    output.delete(0.0, END)
    try:
        defi = my_dict[text];
    except:
        defi = "no word"
    output.insert(END, defi)

Button(window, text = "send", width = 5, command = click).grid(row = 2, column = 0, sticky = W)

# 4 다른 레이블 생성
Label(window, text = "\n 정의:").grid(row = 3, column = 0, sticky = W)

# 5. 텍스트 박스 생성
output = Text(window, width = 75, height = 6, wrap = WORD, background = "light green")
output.grid(row=4, column = 0, columnspan = 2, sticky = W)
# columnspan 의미는 2개까지 칼럼, 즉 0,1 칼럼의 의미... wrap= WORD는 무엇??
my_dict = {"first" : "it is gut", "second" : 'and i work on it', "third" : 'thanks alot'};

window.mainloop()
```

```python
from tkinter import *
from decimal import *

window = Tk()
window.title("calc");

# 내용 수정이 가능한 엔트리 위젯을 사용해 결과 디스플레이 사용
display = Entry(window, width = 45, bg = "light green");
display.grid()

# 숫자 버튼
def click_one():
    display.insert(END, "1")
Button(window, text = "1", width = 5, command = click_one).grid(row = 1, column = 0)

def click_two():
    display.insert(END, "@");
Button(window, text = "2", width = 5, command = click_two).grid(row=2 , column = 0)

def click_three():
    display.insert(END, "kkk");
Button(window, text="3", width = 5, command = click_three).grid(row=3, column = 0)

window.mainloop()
```

```python
from tkinter import *
from decimal import *


def click(key):
    display.insert(END, key)
    
window = Tk()
window.title("my calc")

top_row = Frame(window) # ??
top_row.grid(row=0, column=0, columnspan=2, sticky = N)

display = Entry(top_row, width = 45, bg = "light green") # 수정 가능한 엔트리 위젯이라 함

# 숫자 버튼 프레임 생성
num_pad = Frame(window)
num_pad.grid(row=1, column = 0, sticky=W)

# 숫자 버튼 프레임에 제공될
num_pad_list = ["7", "8", "9",
               '4', '5', '6',
               '1', '2', '3',
               '0', '.', '='];

# 반복문 숫자 버튼
r = 0;
c = 0;

for btn in num_pad_list:
    Button(num_pad, text = btn, width = 5, command=click(btn)).grid(row = r, column = c)
    c = c+1
    if c > 2:
        c = 0
        r = r+1
        
functions = Frame(window)
functions.grid(row = 3, column = 1, sticky = E)
functions_lst = ['actorial(!)', '-> roman', '->binary', 'binary->10']

k = 0;
m = 0;
for b in functions_lst:
    def cmd(x = b):
        click(x)
    Button(functions, text = b, width = 13, comman = cmd).grid(row = k, column = m)
    k = k+1
window.mainloop()

# 고의적 실수가 있나?? top row가 안나옴
```

```python
def click(key):
    if key == "=":
        result = str(eval(display.get()))
        #eval(expression)은 실행 가능한 문자열(1+2, 'hi' + 'a' 같은 것)을 
        #입력으로 받아 문자열을 실행한 결과값을 리턴하는 함수이다.
        display.insert(END, "=" + result);
    # C버튼 눌러졌을때 
    elif key == "c":
        display.delete(0, END);
    elif key == functions_lst[0]:
        num = display.get();
        ans = 0;
        for i in range(num,1, -1):
            i = i * (i-1)
            ans += i;
        display.insert(END, functions_lst[0], ans)
    # 그 외 다른 키를 눌렀을때
    else:
        display.insert(END, key);

window = Tk();
window.title("calc")

top_row = Frame(window)
top_row.grid(row=0, column = 0, columnspan=2, sticky = N)
display = Entry(top_row, width = 45, bg ="light green")
display.grid(row = 0, column = 0, sticky = N)

# 숫자 버튼 프레임 생성
num_pad = Frame(window)
num_pad.grid(row=1, column = 0, sticky=W)

# 숫자 버튼 프레임에 제공될
num_pad_list = ["7", "8", "9",
               '4', '5', '6',
               '1', '2', '3',
               '0', '.', '='];

# 반복문 숫자 버튼
r = 0;
c = 0;

for btn in num_pad_list:
    def cmd(x = btn):
        click(x);
    Button(num_pad, text = btn, width = 5, command=cmd).grid(row = r, column = c)
    c = c+1
    if c > 2:
        c = 0
        r = r+1
        
# 연산자 프레임
operator = Frame(window);
operator.grid(row =1, column=1, sticky = E)
operator_list = [
'*', '/', '+', '-', '{', '}', 'c'
]

for b in operator_list:
    def cmd(x = b):
        click(x);
    Button(operator, text = b, width = 5, command = cmd).grid(row = r, column = c);
    c += 1;
    if c > 1:
        c = 0;
        r += 1;


functions = Frame(window)
functions.grid(row = 3, column = 1, sticky = E)
functions_lst = ['actorial(!)', '-> roman', '->binary', 'binary->10']

for b in functions_lst:
    def cmd(x = b):
        click(x)
    Button(functions, text = b, width = 13, comman = cmd).grid(row = r, column = c)
    r = r+1





window.mainloop()
```
